```python
def numParticiones(n):
    if n==0:
        print("el numero de particiones para el conjunto de tamaño  0 es 1")
    else:
        #lista base
        l0 = [1]
        #lista siguiente
        l1 = []

        for i in range(n):
            l1.append(l0[-1])
            for j in range(len(l0)):
                l1.append(l1[j]+l0[j])
            #si la sig repeticion se tiene que hacer entonces actualizar lista base y lista siguiete
            if i+1<n:
                l0 = l1
                l1 = []
        print("el numero de particiones para el conjunto de tamaño ",n,"es ", l1[0])


def main():

    for i in range(66):
      numParticiones(i)

main()
```
